- 여러 개의 선을 그리기\
1) plot 함수에 x, y 그래프색과 모양 3개를 묶어서 3개를 표현해서 그리기

```python
import matplotlib.pyplot as plt # matplotlib 라이브러리의 pyplot 모듈을 plt라는 별칭으로 가져옴
import numpy as np # numpy를 np라는 별칭으로 가져옴

t= np.arange(0, 5., 0.2) # np.arange()함수는 지정된 범위에서 일정한 간격으로 값들을 생성
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()

# plt.plot 함수는 x와 y값을 기반으로 여러 데이터를 동시에 그림
# t, t, 'r--' : 첫 번째 그래프는 t에 대해 t를 빨간색 점선(r--)으로 표시
# t, t**2, 'bs' : 두 번째 그래프는 t에 대해 t^2을 파란색 정사각형(bs)로 표시
# t, t**3, 'g^' : 세 번째 그래프는 t에 대해 t^3을 초록색 삼각형(g^)으로 표시
# ~에 대해 ~을 ~로 표시 형식임. (ㅇ,ㅇ,ㅇ)
# plt.show()
```

```python
import matplotlib.pyplot as plt

a = plt.plot([1, 2, 3, 4])
b = plt.ylabel('some numbers')
c = plt.xlabel('some indices')
a += plt.plot([0.4, 1.1, 2.1, 3.4])

print(a) # 이것은 Line2D 객체의 리스트를 출력합니다.
print(b) # plt.ylabel()는 플롯을 수정하며, 반환값은 None입니다.
print(c) # plt.xlabel()는 플롯을 수정하며, 반환값은 None입니다.

plt.show() # 플롯을 화면에 표시

# type과 dir의 올바른 사용
print(type(a[0])) # 첫 번째 Line20 객체의 타입을 출력
print(dir(a[0])) # 첫 번째 Line 객체의 메소드와 속성을 출력
```

- 하나의 캔버스에 두개의 그래프 처리

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
# np.linspace(start, stop, num, endpoint=True) 함수는 start에서 stop까지 num개의 균등한 값을 생성
# 여기서는 -π에서 π까지의 구간을 256개의 점으로 나눠 배열 X를 생성
C, S = np.cos(X), np.sin(X)

plt1 = plt.plot(X, C)
plt2 = plt.plot(X, S)

print (plt1, plt2)
plt.show()
```

- Figure 객체를 두개 만들고 내부에 처리

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

fig1 = plt.figure(1)
plt1 = plt.plot(X, C)

fig2 = plt.figure(2)
plt2 = plt.plot(X, S)

print(plt1, plt2)
print(fig1, fig2)

plt.show()
```

- seaborn을 이용하면 그래프의 격자가 꾸며짐

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn의 스타일 설정
sns.set(style = 'darkgrid') # 'white', 'ticks' 등 다른 스타일도 선택할 수 있음.

# 데이터 생성
x = np.linspace(0, 10, 1000)
y = np.power(x, 2)

# Seaborn을 사용하여 선 플롯 생성
sns.lineplot(x = x, y = y)

# 레이블과 제목 추가
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('2D Plot')

# 플롯 표시
plt.show()
```
