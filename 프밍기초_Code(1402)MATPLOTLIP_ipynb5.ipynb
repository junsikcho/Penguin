# bar 함수

-기본\
1) bar함수는 폭을 0.8, 파란색 막대가 기본으로 처리

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = [20, 35, 30, 35, 27]
ind = np.arange(N)
# ind는 각 막대의 x축 위치를 나타내는 배열로 사용
plt.bar(ind, menMeans)
plt.show()
# bar함수는 폭을 0.8, 파란색 막대가 기본으로 처리
```

- 폭 늘리기\
1) bar함수는 위치와 값을 막대그래프로 표시

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = [20, 35, 30, 35, 27]
ind = np.arange(N)

plt.bar(ind, menMeans, width=1.0, color = 'r')
plt.show()
```

- bar함수는 막대 그래프의 폭을 0.5로 처리

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = [20, 35, 30, 35, 27]
ind = np.arange(N)

plt.bar(ind, menMeans, width = 0.5, color = 'r')
plt.show()
```

- 다중 막대 그래프\
1) bar함수는 막대 그래프의 폭을 0.33로 처리해 이중 막대 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
ind = np.arange(N)
menMeans = [20, 35, 30, 35, 27]
womenMeans = [25, 32, 34, 20, 25]

p1 = plt.bar(ind + 0.0, menMeans, width=0.33, color = 'r')
p2 = plt.bar(ind + 0.33, womenMeans, width = 0.33, color = 'y')
plt.xlabel("The number of people")
plt.ylabel("Scores")
plt.title("Scores by group and genders")
plt.xticks(ind + 0.33/2., ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

1) 수평 막대그래프를 그리기 위해서는 반대방향의 데이터 m.pop 앞에 minus 부호(-)를 부여해야 함

```python
import numpy as np
import matplotlib.pyplot as plt

w_pop = np.array([5, 30, 45, 40])
m_pop = np.array([4, 28, 40, 35])
x = np.arange(4)

a = plt.barh(x, w_pop, color = 'r')
b = plt.barh(x, -m_pop)

plt.show()
```

# hist함수 : 기본

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 5.0
rms = 3.0
n = 1000

data = np.random.normal(mean, rms, n)
plt.hist(data)

plt.xlabel('data')
plt.show()
```

# hist함수 : 범주 나누기

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 5.0
rms = 3.0
n = 1000

data = np.random.normal(mean, rms, n)
plt.hist(data, bins = 30)

plt.xlabel('data')
plt.show()
```

# hist함수 : 내부에 그리기

- 그래프 내에 범주를 재정의해서 그래프를 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 5.0
rms = 3.0
n = 1000

data = np.random.normal(mean, rms, n)

plt.hist(data, facecolor = 'red', alpha = 0.4, histtype = 'stepfilled')

bins = np.arange(-5, 16, 1)
plt.hist(data, bins, histtype = 'stepfilled')

plt.xlabel('data')
plt.show()
```

# hist함수 : 파라미터

- density = true를 사용하면 히스토그램 합이 1, edgecolor = 'black' -> 경계선을 만들기

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000)

num_bins = 20

plt.hist(x, num_bins, density=True, facecolor = 'green',
         alpha=0.5, edgecolor = 'black')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ: $\mu=100$, $\sigma = 15$')

plt.show()
```

# Axes.hist 메소드

- hist 메소드를 이용해서 처리
- Cumulative는 누적 분포를 나타내는 그래프를 추가로 그리기 위한 파라미터

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (6, 3))

ax1.hist(data, bins = 30, density = True,
         color = 'b', edgecolor = 'black')
ax2.hist(data, bins = 30, density = True,
         color = 'r', cumulative = True, edgecolor = 'black')
plt.show()
```
