# 내부 함수, lambda, map()

- 내부 함수 : 함수 안에 함수가 있는 형태

```python
def outFunc(v1, v2) :
  def inFunc(num1, num2) :
    return num1 + num2
  return inFunc(v1, v2)
print(outFunc(10, 20))
```

```python
print(inFunc(10,20))
```

- 람다 함수 : 함수를 한 줄로 간단하게 만들어 줌

```python
def  hap(num1, num2) :
  res = num1 + num2
  return res
print(hap(10, 20))
```

```python
hap2 = lambda num1, num2 : num1 + num2
print(hap2(10, 20))
```

- 매개변수에 기본값을 설정\
1) 매개변수를 지정하지 않으면 기본값으로 설정, 매개변수를 넘겨주면 기본값 무시

```python
hap3 = lambda num1 = 10, num2 = 20 : num1 + num2
print(hap3())
```

```python
print(hap3(100, 200))
```

- 리스트에 모두 10을 더하는 코드

```python
# Sol 1
myList = [1, 2, 3, 4, 5]
def add10(num) :
  return num + 10

for i in range(len(myList)) :
  myList[i] = add10(myList[i])
print(myList)
```

```python
# Sol 2
myList = [1, 2, 3, 4, 5]
add10 = lambda num : num + 10
myList = list(map(add10, myList))
print(myList)
```

- 2행과 3행을 합친 코드

```python
# Sol 3
myList = [1, 2, 3, 4, 5]
myList = list(map(lambda num : num + 10, myList))
print(myList)
```

```python
list1 = [1, 2, 3, 4]
list2 = [10, 20, 30, 40]

haplist = list(map(lambda n1, n2: n1 + n2, list1, list2))

print(haplist)
```

```python
# 입력한 숫자를 1까지 세는 함수를 재귀 함수
def count(num) :
  if num >= 1 :
    print(num, end = ' ')
    count(num - 1)
  else :
    return
count(10)
count(20)
```

# 제너레이터와 yield

- yield 문 : 함수를 종결하지 않으면서 값을 계속 반환

```python
def genFunc() :
  yield 1
  yield 2
  yield 3

print(list(genFunc()))
```

- 제너레이터(Generator : 생성자) 함수 : yield가 포함된 함수

1) yield문으로 값을 반환한 후 계속 진행

```python
def genFunc(num) : # 1행의 genFunc()함수는 매개변수 num에 5를 대입
  for i in range(0, num) : # 2행에서 range()함수로 0 ~ 4를 반복
    yield i # 3행에서 0 반환
    print('제너레이터 진행 중')
for data in genFunc(5) : # 5행에서 for 문이 실행될 때 genFunc()함수 호출
  print(data) # 6행에서 0 출력
```
