# imshow()함수

- imshow()함수를 이용해서 이미지 출력 colorbar함수를 이용옆에 옆에 colorbar를 출력

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

A = np.random.random((100, 100))
plt.imshow(A)
plt.hot()
plt.colorbar()
plt.show()
```

# image.read 함수

- 이미지 파일을 읽고 이를 ndarray로 전환해서 imshow함수로 그래프 출력

# text함수 : text 붙이기

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000)

num_bins = 20

plt.hist(x, num_bins, density = True, facecolor = 'green',
         alpha = 0.5, edgecolor = 'black')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ: $\mu=100$, $\sigma = 15$')
plt.text(60, 0.025, '$\mu=100, \ \sigma=15$')

plt.show()
```

# annotate 함수 : 기초

- annotate 함수는 문장열, xy(화살표 끝 지시), xytext(문자열 시작 위치), arrowpros(화살표) 그래프에 주석을 표시

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2,1), xytext=(3,1.5),
             arrowprops=dict(facecolor = 'black', shrink = 0.05))
plt.ylim(-2.2)
plt.show()
```

# figure / subplot함수 : 예시

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t) :
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
